# Spectrograms Benchmark
This notebook is used for testing and benchmarking performance of a list/table of spectrograms of various baseline and polarisation.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from time import sleep
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from model.message_topic import MessageTopic
from model.protocol import Protocol
import mock_payload_generator as pg
import utils
from producer import Producer
```

```python
producer = Producer()
```

## Payload Size

```python
channels = list(range(10000, 80000, 10000))
num_spectrograms = list(range(10, 110, 10))
```

```python
data = []
unit = "MB"

for _, num_spectrogram in enumerate(num_spectrograms):
    for _, channel in enumerate(channels):
        _, size = pg.spectrogram_protobuf(num_spectrogram, channel)
        a = {
            "Size" : utils.convert_bytes(size, unit),
            "Spectrogram": num_spectrogram,
            "Channel": channel,
            "Encoding": "ProtoBuf",
        }
        data.append(a)

        _, size = pg.spectrogram_json(num_spectrogram, channel)
        b = {
            "Size" : utils.convert_bytes(size, unit),
            "Spectrogram": num_spectrogram,
            "Channel": channel,
            "Encoding": "JSON",
        }
        data.append(b)
```

### Plot Payload Size

```python
df_size = pd.DataFrame(data)

col_size = df_size["Size"]
max = col_size.max()
min = col_size.min()

# select ProtoBuf rows
df1_size = df_size.loc[df_size["Encoding"] == "ProtoBuf"]
df1_size = df1_size.pivot(index="Channel", columns="Spectrogram", values="Size")
df1_size = df1_size.sort_values(by=['Channel'],  ascending=False)

# select JSON rows
df2_size = df_size.loc[df_size["Encoding"] == "JSON"]
df2_size = df2_size.pivot(index="Channel", columns="Spectrogram", values="Size")
df2_size = df2_size.sort_values(by=['Channel'],  ascending=False)
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
cmap = "BuPu"
```

```python

fig, ax = plt.subplots(ncols=3, gridspec_kw=dict(width_ratios=[4,4,.35]), figsize=(3, 4))

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)

yticklabels = ["{:,.0f}".format(d / 1000) + "K" for d in list(reversed(channels))]

p1 = sns.heatmap(df1_size, cmap=cmap, ax=ax[0], vmin=min, vmax=max, cbar=False, yticklabels=yticklabels)
p2 = sns.heatmap(df2_size, cmap=cmap, ax=ax[1], vmin=min, vmax=max, cbar=False, yticklabels=False)

cb = fig.colorbar(ax[1].collections[0], cax=ax[2], drawedges=False)
cb.outline.set_visible(False)
 
p1.set_xlabel("")
p1.set_ylabel("")
p2.set_xlabel("")
p2.set_ylabel("")

plt.show()
```

## Payload Deserialization and Transmission

Send each payloads to the browser 5 or 10 times and collect deserialisation, and transmisssion time in the browser.

```python
num_iter = 5
```

### ProtoBuf

```python
await producer.start()

for _, num_spectrogram in enumerate(num_spectrograms):
    for idx, channel in enumerate(channels):
        for itr in range(num_iter):
            sleep(10)
            # print(f"itr: {itr}, idx: {idx}, channel: {channel}, num_spectrogram: {num_spectrogram}")
            payload, _ = pg.spectrogram_protobuf(num_spectrogram, channel)
            await producer.produce(payload, f'{Protocol.PROTOBUF}_{MessageTopic.SPECTROGRAMS}')

await producer.stop()
print("Complete!")
```

### Plot Deserialization Time

```python
idx = 0
_df = []

for _, num_spectrogram in enumerate(num_spectrograms):
    for _, channel in enumerate(channels):
        for itr in range(num_iter):
            
            d1 = {
                "Spectrogram": num_spectrogram,
                "Channel": channel,
                "Itr": itr,
                "Time": protobuf_decoding_time[idx],
                "Encoding": "ProtoBuf",
            }
            _df.append(d1)

            d2 = {
                "Spectrogram": num_spectrogram,
                "Channel": channel,
                "Itr": itr,
                "Time": json_decoding_time[idx],
                "Encoding": "JSON",
            }
            _df.append(d2)

            idx += 1

df_des = pd.DataFrame(_df)
```

```python
# https://stackoverflow.com/questions/30328646/python-pandas-group-by-in-group-by-and-average

df_des = df_des.groupby(['Spectrogram', 'Channel', "Encoding"], as_index=False)['Time'].mean()
```

```python
col_size = df_des["Time"]
max = col_size.max()
min = col_size.min()

# select ProtoBuf rows
df_des_pb = df_des.loc[df_des["Encoding"] == "ProtoBuf"]
df_des_pb = df_des_pb.pivot("Channel", "Spectrogram", "Time")
df_des_pb = df_des_pb.sort_values(by=['Channel'],  ascending=False)

# select JSON rows
df_des_json = df_des.loc[df_des["Encoding"] == "JSON"]
df_des_json = df_des_json.pivot("Channel", "Spectrogram", "Time")
df_des_json = df_des_json.sort_values(by=['Channel'],  ascending=False)
```

```python
cmap = "GnBu"

fig, ax = plt.subplots(ncols=3, gridspec_kw=dict(width_ratios=[4,4,.35]), figsize=(3, 4))

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)

yticklabels = ["{:,.0f}".format(d / 1000) + "K" for d in list(reversed(channels))]

p1 = sns.heatmap(df_des_pb, cmap=cmap, ax=ax[0], vmin=min, vmax=max, cbar=False, yticklabels=yticklabels)
p2 = sns.heatmap(df_des_json, cmap=cmap, ax=ax[1], vmin=min, vmax=max, cbar=False, yticklabels=False)

cb = fig.colorbar(ax[1].collections[0], cax=ax[2], drawedges=False)
cb.outline.set_visible(False)
 
p1.set_xlabel("")
p1.set_ylabel("")
p2.set_xlabel("")
p2.set_ylabel("")

plt.show()
```
